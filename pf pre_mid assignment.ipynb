```python
#part B

annual_salary=(float(input("Enter your annual salary")))# ask user his annual salary
portion_saved=(float(input("Enter the percentage of your salary to save,(as a decimal)")))# user will enter percentage he wants to save
total_cost=(float(input("Enter the total cost of your dream house"))) #ask user his total cost of your dream house
semi_annual_raise=(float(input("Enter the semi annual raise,(as a decimal)")))# ask user semi annual raise(increased salary)
#writing down all values so it can help to code better
r=0.04 #annual intrest
portion_down_payment=0.25

down_payment=total_cost*portion_down_payment #down payment formula

current_saving=0 #initializing current_saving to 0 as user does not have any saving now

monthly_salary=annual_salary/12 #tells about user monthly salary
months=0 #initializing month to 0 as user have not started saving yet


while current_saving<down_payment:#used loop to calculate months as there are more than 1 month 
    
   #formulas 
    monthly_saving=monthly_salary*portion_saved  #calculates monthly saving
    monthly_intrest=current_saving*(r/12)  #calculates monthly intrest(which user gets by investing money)
    current_saving=current_saving+monthly_saving+monthly_intrest #calculates cuurent saving for every month till user collects money for house

    months+=1 #increments month as there are more than 1 month

    if months%6==0: #condition for  semi annual raise
      annual_salary=annual_salary*(1+semi_annual_raise)  #as annual salary increases,formula changes
      monthly_salary=annual_salary/12 #monthly salary also changes

print(f"Number of months: {months}") #finally tells user about  total months  to collect money
```
